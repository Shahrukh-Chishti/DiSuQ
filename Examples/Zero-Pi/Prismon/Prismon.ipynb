```python
from DiSuQ.Torch import models
from torch import tensor
from numpy import arange,linspace,meshgrid,array
from DiSuQ.utils import plotCompare,plotHeatmap
from torch import set_num_threads
set_num_threads(32)
```

```python
Na,Nc  = 25,25
PhiA = linspace(0,1,Na,endpoint=True)
PhiC = linspace(0,1,Nc,endpoint=True)
flux_range = meshgrid(PhiA,PhiC)
flux_range = array((flux_range[0].flatten(),flux_range[1].flatten())).T
```

```python
flux_range.shape
```

```python
basis = [5,5,5,5,5]
basis = {'O':[5,5,5],'J':[5,5],'I':[]}
flux_point = ['La','Lc']
flux_manifold = [flux for flux in tensor(flux_range)]
circuit = models.prismon(basis,sparse=False)
```

```python
circuit.modeDistribution()
```

```python
H_LC = circuit.kermanHamiltonianLC()
H_J = circuit.kermanHamiltonianJosephson
```

```python
H_LC.shape
```

```python
E0,(Ex1,Ex2) = circuit.spectrumManifold(flux_point,flux_manifold,H_LC,H_J,excitation=[1,2],grad=False,log=True)
```
